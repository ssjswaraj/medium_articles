```python
from pyspark.sql import SparkSession
from delta import configure_spark_with_delta_pip
builder = (
    SparkSession.builder
    .appName("DeltaLakeApp")
    .config("spark.sql.extensions",
            "io.delta.sql.DeltaSparkSessionExtension")
    .config("spark.sql.catalog.spark_catalog",
            "org.apache.spark.sql.delta.catalog.DeltaCatalog")
)

spark = configure_spark_with_delta_pip(builder).getOrCreate()
```

```python
import importlib.metadata
print("spark version : ",spark.version)
print("delta version : ",importlib.metadata.version("delta-spark"))
```

# 2. start from here

## 2.1 create database

```python
spark.sql("create database deltahub")
spark.sql("show databases").show()
```

```python
from delta.tables import *
from pyspark.sql.types import StructType, StructField, IntegerType, StringType
```

## 2.2 create table

```python
spark.sql("""
   create table deltahub.employee(
    id int,
    name string,
    gender string,
    salary int,
    department string
   )
   using delta
   location '/content/spark-warehouse/deltahub.db/employee'
""")
```

## 2.3 Insert

```python
spark.sql("""
  insert into deltahub.employee
  values(1,"bruce","m",1000,"it")
""")
spark.sql("""select * from deltahub.employee""").show()
```
